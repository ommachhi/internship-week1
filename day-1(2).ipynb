# Print Even or Odd Numbers Using(if–else)

```python
num = int(input("Enter Number:"))
if(num % 2 == 0):
    print(f"{num} is a Even Number")
else:
    print(f"{num} is a Odd Number")
```

# Largest of Three Numbers (if–elif–else)

```python
n1 = int(input("Enter number 1:"))
n2 = int(input("Enter number 2:"))
n3 = int(input("Enter number 3:"))

if(n1>n2 and n1>n3):
    print(f"{n1} is Largest Number")
elif(n2>n1 and n2>n3):
    print(f"{n2} is Largest Number")
elif(n3>n1 and n3>n2):
    print(f"{n3} is a Largest Number")
else:
    print("All The Numbers Are Same")
```

# Sum of Numbers from 1 to N (for loop)

```python
n = int(input("Enter Number:"))
tot = 0
for i in range(1,n+1):
    tot +=i
print(f"sum of numbers is: {tot}")
```

# Check Prime Number (Function)

```python
def prime(n):
    for i in range(2,n):
        if n % i == 0:
            return False
        return True
n = int(input("Enter Number:"))
if(prime(n)):
    print("prime number")
else:
    print("not a prime")
```
