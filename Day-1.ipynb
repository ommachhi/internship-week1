# Data types

```python
name = "om"
age = 12
number = 12.3
data = True

print(f"type of name is {type(name)}")
print(f"type of age is {type(age)}")
print(f"type of number is {type(number)}")
print(f"type of data is {type(data)}")
```

## Program 1: Simple Calculator
This program takes two numbers from the user and performs arithmetic operations.

```python
num1 = int(input("Enter Number 1:"))
num2 = int(input("Enter Number 2:"))

print(f"sum of {num1} and {num2} is:{num1+num2}")
print(f"sub of {num1} and {num2} is:{num1-num2}")
print(f"mul of {num1} and {num2} is:{num1*num2}")
print(f"div of {num1} and {num2} is:{num1/num2}")

```
