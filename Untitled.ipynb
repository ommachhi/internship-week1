```python
class BankAccount:
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance

    def display_balance(self):
        return f"{self.name}, Your balance is {self.balance}"
```

```python
def deposit(account, amount):
    try:
        if amount <= 0:
            raise ValueError("Amount must be positive")
        account.balance += amount
        return f"Deposited {amount}"
    except ValueError as e:
        return f"Error: {e}"


def withdraw(account, amount):
    try:
        if amount > account.balance:
            raise ValueError("Insufficient balance")
        account.balance -= amount
        return f"Withdrawn {amount}"
    except ValueError as e:
        return f"Error: {e}"
```

```python
from bank.account import BankAccount
from bank.operations import deposit, withdraw

# Create account
acc = BankAccount("Om", 1000)

print(acc.display_balance())

# Deposit
print(deposit(acc, 500))

# Withdraw
print(withdraw(acc, 2000))  # error case

# Final balance
print(acc.display_balance())

try:
    amount = int(input("Enter deposit amount: "))
    print(deposit(acc, amount))
except ValueError:
    print("Invalid input")
```
